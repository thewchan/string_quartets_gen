```python
def gen_piano_roll(file_path):
    midi = pretty_midi.PrettyMIDI(file_path)
    (denominator, numerator) = (midi.time_signature_changes[0].denominator, midi.time_signature_changes[0].numerator)
    sixteenth_factor = numerator / denominator * 16
    sixteenth_length = midi.get_downbeats()[1] / sixteenth_factor
    return np.array([
        np.argmax(midi.instruments[part_num].get_piano_roll(fs=(1.0/sixteenth_length)), axis=0) for part_num in range(4)
    ], dtype=object).T

gen_piano_roll("../data/midis/3. What's This.mscx.mxl.midi")
```

```python
data_container = np.empty(len(os.listdir('../data/midis')), dtype=object)


for num, file in enumerate(os.listdir('../data/midis')):
    print(f'Converting file #{num} {file} to numpy array...', end = '\t\t')
    try:
        data_container[num] = gen_piano_roll('../data/midis/' + file)
        print('Done!')
    except Exception as e:
        print(f'Error converting: {e.__class__.__name__} - {e}, skipping...')
    
```

```python
import numpy as np

data = np.load('../data/stringquartet16thseparated.npz', allow_pickle=True)
```
