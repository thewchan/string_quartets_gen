```python
quartet_parts = ['Violin I', 'Violin II', 'Viola', 'Violoncello']
quartet_files = []

for music_file in music_files:
    try:
        if [parts.partName for parts in music21.converter.parse('../mxl/' + music_file).parts] == quartet_parts:
            quartet_files.append(music_file)
            print(f'Adding {music_file} to quartet muisc list.')
        else:
            print(f'Skipping {music_file}.')
    except (music21.note.NotRestException, TypeError, music21.musicxml.xmlToM21.MusicXMLImportException, music21.harmony.HarmonyException) as e:
        print(f'Parsing error: {e}; skipping {music_file}.')          
    
```

```python
import dill

with open('../quartet_files.pkd', 'wb') as f:
    dill.dump(quartet_files, f)
```

```python
import subprocess


mxl_list = os.listdir('../mxl')

for file in mxl_list:
    if file in quartet_files:
        print(f'{file} is a quartet; moving to data directory.')
        subprocess.run(['mv', f'../mxl/{file}', f'../quartet_raw/{file}'])
    else:
        print(f'{file} is not a quartet; skipping.')
```
